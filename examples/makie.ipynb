# Visualization with Makie

Tangles.jl has support for visualization of Tensor Networks using the GraphMakie.jl library.

```julia
using Tangles
using GraphMakie
using CairoMakie
using NetworkLayout
using Random
```

```julia
CairoMakie.activate!(type = "svg")
Random.seed!(0)
```

Any type inheriting `AbstractTensorNetwork` automatically implements the `GraphMakie.graphplot` method.

```julia
tn = rand(SimpleTensorNetwork, 30, 3)
graphplot(tn; layout=Stress())
```
